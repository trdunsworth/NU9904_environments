```python
adftest_mcod = adfuller(day['mco'], autolag='AIC', regression='ct')
print("ADF Test Results")
print("Null Hypothesis: The series has a unit root (non-stationary)")
print("ADF-Statistic:", adftest_mcod[0])
print("P-Value:", adftest_mcod[1])
print("Number of lags:", adftest_mcod[2])
print("Number of observations:", adftest_mcod[3])
print("Critical Values:", adftest_mcod[4])
print("Note: If P-Value is smaller than 0.05, we reject the null hypothesis and the series is stationary")
```

```python
adftest_mcod2 = adfuller(day_c['mco'], autolag='AIC', regression='ct')
print("ADF Test Results")
print("Null Hypothesis: The series has a unit root (non-stationary)")
print("ADF-Statistic:", adftest_mcod2[0])
print("P-Value:", adftest_mcod2[1])
print("Number of lags:", adftest_mcod2[2])
print("Number of observations:", adftest_mcod2[3])
print("Critical Values:", adftest_mcod2[4])
print("Note: If P-Value is smaller than 0.05, we reject the null hypothesis and the series is stationary")
```

```python
mcod = pd.DataFrame(day, columns=['ds', 'mco'])
tr_mcod = mcod[:365]
tst_mcod = mcod[365:]
```

```python
plt.plot(tr_mcod)
plt.plot(tst_mcod)
plt.title = ("Test and Training Split of Daily Data for Monterey County, CA")
plt.xlabel("Date")
plt.ylabel("Call Count")
plt.show()
```

```python
tr_mcod_model1 = pmd.auto_arima(tr_mcod['mco'],
                      m=7,               # frequency of series
                      seasonal=True,  # TRUE if seasonal series
                      d=None,             # let model determine 'd'
                      test='adf',         # use adftest to find optimal 'd'
                      start_p=0, start_q=0, # minimum p and q
                      max_p=12, max_q=12, # maximum p and q
                      D=None,             # let model determine 'D'
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)

# print model summary
print(tr_mcod_model1.summary())
```

```python
mcod_pred1 = pd.DataFrame(tr_mcod_model1.predict(n_periods=len(tst_mcod)), index=tst_mcod.index)
mcod_pred1.columns = ['predicted_volume']
mcod_pred1
```

```python
plt.figure(figsize=(22,10))
plt.plot(tr_mcod, label="Training")
plt.plot(tst_mcod, label = "Test")
plt.plot(mcod_pred1, label = "Prediction")
plt.legend(loc="upper left")
plt.savefig("mcod_arima.png")
plt.show()
```

```python
mcod_mae = mean_absolute_error(tst_mcod['mco'], mcod_pred1)
mcod_mae
```

```python
mcod_mse = mean_squared_error(tst_mcod['mco'], mcod_pred1)
mcod_mse
```

```python
mcod_r2 = r2_score(tst_mcod['mco'], mcod_pred1)
mcod_r2
```
